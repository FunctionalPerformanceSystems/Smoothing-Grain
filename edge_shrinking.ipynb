```python
import json
import plotly.graph_objects as go
import numpy as np
```

```python
with open("all_edges_n100-id1.json",'r') as f:
    edges = json.load(f)
```

```python
edges['1']
```

```python
# Create traces for points and lines
traces = []

for edge in edges['1']:
    x, y, z = zip(*edge)
    trace_lines = go.Scatter3d(x=x, y=y, z=z, mode='lines', line=dict(color='red', width=2))
    traces.append(trace_lines)

# Create the figure
fig = go.Figure(data=traces)

# Set layout options
fig.update_layout(scene=dict(aspectmode='data'))

# Show the interactive plot
fig.show()
```
