```python
import json
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
```

```python
with open("n100-id1.json",'r') as f:
    coords = json.load(f)
```

```python
coords['1']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot each point and connect them
for points in coords['1']:
    x, y, z = zip(*points)
    ax.scatter(x, y, z, color='b')  # Plot points
    ax.plot(x, y, z, color='r')  # Connect points with lines

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```

```python
# Create traces for points and lines
traces = []
for points in coords['1']:
    x, y, z = zip(*points)
    trace_points = go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(size=5))
    #trace_lines = go.Scatter3d(x=x, y=y, z=z, mode='lines', line=dict(color='red', width=2))
    traces.extend([trace_points])#, trace_lines])

# Create the figure
fig = go.Figure(data=traces)

# Set layout options
fig.update_layout(scene=dict(aspectmode='data'))

# Show the interactive plot
fig.show()
```
