```python
def face_mean(numeral: str, face_num: int):
    same_coords = []
    mean_coord = []

    for j in range(0,3):
        for i in range(0, len(coords[numeral][face_num])):
            same_coords.append(coords[numeral][face_num][i][j])
        mean_coord.append(sum(same_coords)/len(coords[numeral][face_num]))
        same_coords = []

    return np.array(mean_coord)
```

```python
def normals_for_faces(numeral: str):
    coords_for_normal = []
    for i in range(0, len(coords[numeral])):
        coords_for_normal.append(coords[numeral][i][0:3])
    
    normals = []
    merged = []
    for i in range(0, len(coords_for_normal)):
        point1 = np.array(coords_for_normal[i][0])
        point2 = np.array(coords_for_normal[i][1])
        point3 = np.array(coords_for_normal[i][2])

        vector1 = point2 - point1
        vector2 = point3 - point1

        normal_vector = np.cross(vector1, vector2)

        merged_array = np.concatenate((face_mean(numeral, i), normal_vector))
        merged.append(merged_array)

        normals.append(normal_vector)

    return merged
```

```python
def mean_poly(numeral):
    flattened_array = [item for sublist in coords[numeral] for item in sublist]

    # Remove duplicates
    unique_array = [list(t) for t in {tuple(element) for element in flattened_array}]

    same_coords = []
    mean_coord = []

    for j in range(0,3):
        for i in range(0, len(unique_array)):
            same_coords.append(unique_array[i][j])
        mean_coord.append(sum(same_coords)/len(unique_array))
        same_coords = []
    
    return mean_coord
```

```python
def normals_inside(face_num, poly_centre):

    centre_vector = normals_for_faces('1')[face_num][0:3] - poly_centre

    dot_product = np.dot(normals_for_faces('1')[face_num][3:6], centre_vector)

    if dot_product > 0:
        return np.concatenate((normals_for_faces('1')[face_num][0:3],-normals_for_faces('1')[face_num][3:6]))
    else:
        return np.concatenate((normals_for_faces('1')[face_num][0:3], normals_for_faces('1')[face_num][3:6]))
```
