```python
import json
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
```

```python
with open("n100-id1.json",'r') as f:
    coords = json.load(f)
```

```python
with open("all_edges_n100-id1.json",'r') as f:
    edges = json.load(f)
```

```python
import plotly.graph_objects as go

# Create traces for points and lines
traces = []
for points in coords['1']:
    x, y, z = zip(*points)
    trace_points = go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(size=5))
    traces.append(trace_points)

for edge in edges['1']:
    x, y, z = zip(*edge)
    trace_lines = go.Scatter3d(x=x, y=y, z=z, mode='lines', line=dict(color='red', width=2))
    traces.append(trace_lines)

# Create the figure
fig = go.Figure(data=traces)

# Set layout options
fig.update_layout(scene=dict(aspectmode='data'))

# Show the interactive plot
fig.show()
```

```python
import plotly.graph_objects as go

# Create traces for points and lines
traces = []
unique_lines = set()

for points in coords['1']:
    x, y, z = zip(*points)
    trace_points = go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(size=5))
    traces.append(trace_points)

    for edge in edges['1']:
        edge_tuple = tuple(map(tuple, edge))  # Convert inner lists to tuples
        if edge_tuple not in unique_lines:
            x, y, z = zip(*edge_tuple)
            trace_lines = go.Scatter3d(x=x, y=y, z=z, mode='lines', line=dict(color='red', width=2))
            traces.append(trace_lines)
            unique_lines.add(edge_tuple)

# Create the figure
fig = go.Figure(data=traces)

# Set layout options
fig.update_layout(scene=dict(aspectmode='data'))

# Show the interactive plot
fig.show()
```
