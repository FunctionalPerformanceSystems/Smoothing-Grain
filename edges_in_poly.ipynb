```python
import json
```

```python
poly_dict = {}
with open("n100-id1.tess", "r") as file:
    lines = file.readlines()
    for i, line in enumerate(lines):
        if '**polyhedron' in line:
            num_poly = int(lines[i+1])
            for j in range(i+2, i+2+num_poly):
                poly_info = lines[j].split()
                poly_dict[int(poly_info[0])] = [abs(int(elem)) for elem in poly_info[2:int(poly_info[1])+2]]
```

```python
print(poly_dict)
```

```python
face_dict = {}
with open("n100-id1.tess", "r") as file:
    lines = file.readlines()
    for i, line in enumerate(lines):
        if '**face' in line:
            num_faces = int(lines[i+1])
            for j in range(i+2, i+2+4*num_faces, 4):
                faces_info = lines[j].split()
                face_dict[int(faces_info[0])] = [int(elem) for elem in faces_info[2:int(faces_info[1])+2]]
```

```python
print(face_dict)
```

```python
edges_dict = {}
with open("n100-id1.tess", "r") as file:
    lines = file.readlines()
    for i, line in enumerate(lines):
        if '**edge' in line:
            num_edge = int(lines[i+1])
            for j in range(i+2, i+2+num_edge):
                edge_info = lines[j].split()
                edges_dict[int(edge_info[0])] = [abs(int(elem)) for elem in edge_info[1:3]]
```

```python
print(edges_dict)
```

```python
vertex_dict = {}
with open("n100-id1.tess", "r") as file:
    lines = file.readlines()

    for i, line in enumerate(lines):
        if '**vertex' in line:
            num_vertices = int(lines[i+1])
            for j in range(i+2, i+2+num_vertices):
                vertex_info = lines[j].split()
                vertex_dict[int(vertex_info[0])] = [float(coord) for coord in vertex_info[1:4]]
```

```python
print(vertex_dict)
```

```python
poly_faces = {}

for key, value in poly_dict.items():
    poly_faces[key] = [face_dict[num] for num in value if num in poly_dict[key]]

print(poly_faces)
```

```python
edges_for_poly_faces = {}

for key, value in poly_faces.items():
    #edges_for_poly_faces[key] = [edges_dict[num] for num in value if num in poly_faces[key]]
    edges_for_poly_faces[key] = [[edges_dict[num] for num in sublist] for sublist in value]

print(edges_for_poly_faces)
```

```python
edges_in_poly = {}

for key, value in edges_for_poly_faces.items():
    #edges_for_poly_faces[key] = [edges_dict[num] for num in value if num in poly_faces[key]]
    #edges_for_poly_faces[key] = [[edges_dict[num] for num in sublist] for sublist in value]
    edges_in_poly[key] = [[[vertex_dict[num] for num in sublist2 ] for sublist2 in sublist] for sublist in value]

print(edges_in_poly)
```

```python
with open("edges_in_poly.json",'w') as f:
    json.dump(edges_in_poly, f)
```

```python
edges_for_face = {}
with open("n100-id1.tess", "r") as file:
    lines = file.readlines()
    for i, line in enumerate(lines):
        if '**face' in line:
            num_faces = int(lines[i+1])
            for j in range(i+3, i+3+4*num_faces, 4):
                edge_face_num = lines[j-1].split()
                edge_face_info = lines[j].split()
                edges_for_face[edge_face_num[0]] = [abs(int(elem)) for elem in edge_face_info[1:int(edge_face_info[0])+1]]
```

```python
print(edges_for_face)
```
